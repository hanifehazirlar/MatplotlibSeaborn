```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("ssc_2015-21.xlsx")
df
```

```python
df.info()
```

```python
df["Yıl"] = df["Yıl"].astype(str)
```

```python
#df['Yıl'] = pd.to_datetime(df['Yıl'])
```

```python
df.info()
```

```python
yıllara_göre_toplam_suç = df.groupby("Yıl")["Toplam"].sum()
yıllara_göre_toplam_suç
```

```python
plt.plot(yıllara_göre_toplam_suç.index, yıllara_göre_toplam_suç.values)
plt.xlabel('Yıl')
plt.ylabel('Toplam')
plt.title('yıllara_göre_toplam_suç')
plt.xticks(rotation=45)
plt.grid(True);
```

```python
sns.lineplot(data=df, x='Yıl', y='Toplam', hue='Cinsiyet')
plt.xlabel('Yıl')
plt.ylabel('Toplam')
plt.title('Yıllara ve Cinsiyete Göre Toplam Suçlar')
plt.xticks(rotation=45)
plt.grid(True);
```

```python
cinsiyete_göre_toplam_suç = df.groupby("Cinsiyet")["Toplam"].sum()
cinsiyete_göre_toplam_suç
```

```python
plt.bar(cinsiyete_göre_toplam_suç.index, cinsiyete_göre_toplam_suç)
plt.xlabel('Cinsiyet')
plt.ylabel('Toplam Suç Sayısı')
plt.title('Cinsiyete Göre Toplam Suç');
```

```python
numeric= df.select_dtypes(include='int64')
numeric
```

```python
df = df.drop('Toplam', axis=1, inplace=True)
df
```

```python
suç_sıralaması= numeric.sum().sort_values()
suç_sıralaması
```

```python
plt.figure.figsize=(15,10)
suç_sıralaması.plot(kind='barh')
plt.ylabel('Suç Türleri')
plt.xlabel('Toplam')
plt.title('Suç Türleri Sıralaması')
plt.show()
```

```python
en_yüksek_suç_türleri= suç_sıralaması.nlargest(6)
en_yüksek_suç_türleri
```

```python
plt.pie(en_yüksek_suç_türleri, labels=en_yüksek_suç_türleri.index, autopct='%1.1f%%')
plt.title('En Çok İşlenen Suç Türleri')
plt.show()
```
