## 1. Veri setini pandas kütüphanesi kullanarak yükleyin ve index sütununu filmin başlığı yapın

```python
import pandas as pd
```

```python
df = pd.read_csv("IMDB-Movie-Data.csv")
df.head()
```

```python
# let's make the index column the title of the movie 
df.set_index('Title', inplace=True)
df.head(3)
```

## 2. Veri setini tanıyın

```python
# number of rows and columns
print("Number of rows:   ",df.shape[0])
print("Number of columns:",df.shape[1])
```

```python
# information about data
df.info()
```

```python
# describe numerical basic statistics of data
df.describe().T
```

## 3. Veri setindeki sütun isimlerini kullanışlı hale getirin

```python
df.columns
```

```python
# Let's remove the spaces in the column names
df.rename(columns={"Runtime (Minutes)":"Runtime_Minutes",
                   "Revenue (Millions)": "Revenue_Millions"}, inplace=True)
```

```python
df.tail(3)
```

## 4. Missing Value tespit edilsin ve bu kapsamda aşağıdaki şekilde mücadele edilsin.

#### a.Missing value kontrolü hem pandas hemde missingno kütüphanelerini kullanarak yapın

```python
# Missing data

df.isnull().sum()
```

```python
# missing value with percentage

for i in df.columns:
    null_rate = df[i].isnull().sum() / len(df) * 100 
    if null_rate > 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

```python
# Missing value graph with missingno

import missingno as msno 

msno.bar(df);
```

```python
# Missing value graph with seaborn

import seaborn as sns
sns.heatmap(df.isnull(),
            yticklabels=False,
            cbar=False,
            cmap='magma');
```

#### b.Yüksek missing value olan değişken içindeki veriler o değişkenin ortalama skoru ile doldurulsun.

```python
# Column names containing missing values

null_col = df.columns[df.isnull().sum() > 0]

print("Columns containing missing values: ", null_col)
```

```python
# Column with high missing value is Revenue_Millions

df["Revenue_Millions"] = df["Revenue_Millions"].fillna(df["Revenue_Millions"].mean())
```

#### c. Düşük missing value olan değişkendeki gözlemler kaldırılsın.

```python
# There is only one remaining column that can be deleted with dropna

df.dropna(inplace = True)
```

#### d. Missing value kontrolü tekrar hem pandas hemde missingno kütüphanelerini kullanarak kontrol edin ve bu data setini yeni bir değişkene atayın.

```python
# Control missing values

df.isnull().sum()
```

```python
msno.bar(df);
```

```python
# assign the data set to the new variable
dff = df.copy()
dff.head(3)
```

## 5. Temizlenmiş veri seti içinden en yüksek rating e sahip 250 filmi seçerek "top250" isminde bir data seti oluşturun ve bunu lokalinize bu isimle kaydedin. Ardından bu yeni data setini tekrar pandas kütüphanesini kullanarak okuyun.

```python
# Let's choose the 250 movies with the highest rating

top250 = dff.sort_values(by="Rating", ascending=False).head(250)

# let's save this output to local

top250.to_csv('top250.csv')
```

```python
top250.shape
```

```python
# let's read the new data set

df_new = pd.read_csv("top250.csv")
df_new.set_index('Title', inplace=True)
df_new.head(3)
```

## 8. Verileri görselleştirme:

#### a. Matplotlib kütüphanesi kullanarak, yıllara göre filmlerin sayısını gösteren bir çizgi grafik çizin.

```python
import matplotlib.pyplot as plt
```

```python
df_new.Year.value_counts().sort_index()
```

```python
years = df_new['Year'].value_counts().sort_index()

plt.figure(facecolor = "orange",figsize = (8,5))
plt.plot(years.index, 
         years.values,
         marker = "o", 
         color = "red", 
         linestyle = "--",
         markerfacecolor = "yellow")
plt.title("Number of Movies by Year(top250)", fontsize = 16)
plt.show()
```

#### b. Seaborn kütüphanesi kullanarak, en çok film yapılan 10 türü bar grafiğiyle gösterin.

```python
import seaborn as sns
```

```python
# separates movie genres by comma
genres = df_new['Genre'].str.split(',')

# explode method puts each type on a separate line
genres = genres.explode()

# top 10 most recurring movie genres
genres.value_counts().head(10)
```

```python
plt.figure(facecolor = "orange",figsize = (12, 6))
ax= sns.barplot(y = genres.value_counts().head(10).index, x = genres.value_counts().head(10).values)

ax.set(xlabel='Count',ylabel='Genre', title='Top 10 Most Filmed Genres')
ax.set_facecolor("xkcd:mint green")
ax.bar_label(ax.containers[0])

plt.show()
```

#### c. Seaborn ve Plotly kütüphanesi kullanarak, film türlerine göre filmlerin sayısını gösteren iki ayrı pasta grafik çizin.

```python
genres.value_counts()
```

```python
# matplotlib

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
sns.color_palette("pastel")
plt.pie(x=genres.value_counts().values, labels=genres.value_counts().index, autopct='%1.1f%%')
plt.title('Number of Movies by Movie Genre - matplotlib')
plt.show()
```

```python
import plotly.express as px
```

```python
# plotly

fig = px.pie(names=genres.value_counts().index, values=genres.value_counts().values, title='Number of Movies by Movie Genre - Plotly')
fig.update_layout(height=800, width=800, legend=dict(x = 1.2, y = 0.5))
fig.show()
```

## 9. İlişkileri görselleştirme

#### a. Seaborn kütüphanesi kullanarak, veri setindeki nümerik değişkenlerin korelasyonunu gösteren bir heatmap çizin.

```python
df_new.head(3)
```

```python
sns.heatmap(df_new.corr(), annot = True, cmap='YlGnBu');
```

#### b. Seaborn ile oyunculardan birden fazla filmde rol alan oyuncuların rol aldıkları film sayılarına göre ilk 20 oyucuyu görselleştir.  

#### Not: Bu soru için kullanılacak data setimiz 936 satırlık data seti olacak.

```python
df.shape
```

```python
df.head(3)
```

```python
# separates movie genres by comma
actors = df_new['Actors'].str.split(',')

# we need to remove spaces before or after the name
actors = actors.apply(lambda x: [i.strip() for i in x])  

# explode method puts each type on a separate line
actors = actors.explode()

# the top 20 actors according to the number of movies they starred in
actors.value_counts().head(20)
```

```python
plt.figure(facecolor = "orange",figsize = (12, 6))
ax= sns.barplot(y = actors.value_counts().head(20).index, x = actors.value_counts().head(20).values)

ax.set(xlabel='Count',ylabel='Actors', title='Top 20 Actors by Number of Movies')
ax.set_facecolor("xkcd:mint green")
ax.bar_label(ax.containers[0])

plt.show()
```

### Exxtra -- En fazla filmi olan oyuncuların oynadıkları film üzerinden rating ortalamaları

```python
# Average rating for movies starring Robert Downey Jr.
df_new[df_new.Actors.str.contains("Robert Downey Jr.")]["Rating"].mean()
```

```python
# Average rating for movies starring Brad Pitt
df_new[df_new.Actors.str.contains("Brad Pitt")]["Rating"].mean()
```

* We have 2 actors who played in the most movies (8 movies)

* The purpose of this analysis was to question the reason why these actors play in so many films. Considering that their average ratings are not low, it can be said that there are actors who contribute to the ratings and are appreciated.

#### c. Plotly kütüphanesi kullanarak, Popüler 250 Filmin rating puanlarına göre hasılatını buble chart ile gösterin .

```python
df_new.head(3)
```

```python
# Let's select the required columns
bubble_data = df_new[['Rating', 'Revenue_Millions', 'Rank']]

fig = px.scatter(bubble_data, x='Rating', y='Revenue_Millions', size='Revenue_Millions', 
                 hover_name='Rank', title='Revenue of 250 Popular Movies by Rating Scores')
fig.update_layout(width=1100,height=600)

fig.show()
```

```python
# we get the title column with reset_index to see the movie names in the bubble chart 

df_new.reset_index(inplace = True)

# Let's select the required columns
bubble_data = df_new[['Rating', 'Revenue_Millions', 'Title']]

fig = px.scatter(bubble_data, x='Rating', y='Revenue_Millions', size='Revenue_Millions', 
                 hover_name='Title', title='Revenue of 250 Popular Movies by Rating Scores ')
fig.update_layout(width=1100,height=600)

fig.show()
```
