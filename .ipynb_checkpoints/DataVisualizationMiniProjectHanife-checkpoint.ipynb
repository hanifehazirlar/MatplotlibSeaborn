# Proje Adı: Film İncelemesi

## Projenin amacı:

Bu proje ile mini bir EDA (Exploratory Data Analysis) ve farklı veri görselleştirme kütüphanaleri kullanarak film verilerini incelemek ve farklı filmler arasındaki ilişkileri keşfet amaçlanmıştır.

## Veri Seti:

Araştırmada 2006-2016 yılları arasındaki yapılmış filmlerin yer aldığı "IMDB-Movie-Data"  veri seti kullanılacaktır. Veri seti kısaca yıllara göre IMDB'de en yüksek puan alan filmleri içermektedir. Data setinde her film için aşağıdaki değişkenler yer almaktadır:

    Rank --> derece
    Title --> Filmin adı
    Genre --> Filmin türleri
    Description -->Filmin kısa bir olay örgüsü özeti veya açıklaması
    Director -->Filmin yönetmen(ler)i
    Actors --> Filmin ana oyuncu kadrosu
    Year --> Filmin vizyona girdiği yıl
    Runtime (Minutes)-->Filmin dakika cinsinden süresi
    Rating --> Derecelendirme: Filmin IMDB kullanıcıları tarafından verilen ortalama puanı (0-10 arası bir ölçekte)
    Votes -->Filmin aldığı oy sayısı
    Revenue (Millions) --> Filmin dünya çapında ürettiği brüt gelir (ABD doları olarak)
    Metascore -->Filmin Metacritic web sitesinde profesyonel eleştirmenler tarafından verilen ortalama puanı (0-100 arası bir ölçekte)

## 1. Veri setini pandas kütüphanesi kullanarak yükleyin ve index sütununu filmin başlığı yapın

## 4. Missing Value tespit edilsin ve bu kapsamda aşağıdaki şekilde mücadele edilsin.

#### a.Missing value kontrolü hem pandas hemde missingno kütüphanelerini kullanarak yapın

```python
# Control missing values
movies.isnull().sum()
```

```python
#!pip install missingno
```

```python
import missingno as msno
```

```python
# Visualize missing data
msno.matrix(movies);
```

```python
# Visualizing the number of missing data in features
msno.bar(movies);
```

#### b.Yüksek missing value olan değişken içindeki veriler o değişkenin ortalama skoru ile doldurulsun.

```python
#Fill the data in the variable with high missing value with the average score of that variable.
movies['revenue'].fillna(movies['revenue'].mean(),inplace=True)
```

#### c. Düşük missing value olan değişkendeki gözlemler kaldırılsın.

```python
# Remove the observations in the variable with low missing value.
movies.dropna(inplace=True)
```

```python
movies.shape
```

## 5. Temizlenmiş veri seti içinden en yüksek rating e sahip 250 filmi seçerek "top250" isminde bir data seti oluşturun ve bunu lokalinize bu isimle kaydedin. Ardından bu yeni data setini tekrar pandas kütüphanesini kullanarak okuyun.

```python
# Create a data set named "top250" by selecting the 250 movies with the highest rating
top250=movies_new.sort_values(by="rating",ascending=False).head(250)
top250.to_csv("top250.csv")
```

## 8. Verileri görselleştirme:

#### a. Matplotlib kütüphanesi kullanarak, yıllara göre filmlerin sayısını gösteren bir çizgi grafik çizin.

```python
# Import pyplot module from matplotlib library as plt
import matplotlib.pyplot as plt
```

```python
# Unique years sorted by value index
df_top250['year'].value_counts().sort_index()
```

```python
# Unique years sorted by value index and assigned to new variable
years = df_top250['year'].value_counts().sort_index()
# drawn a line chart showing the number of movies by year
plt.figure(facecolor = "pink",figsize = (9,5))
plt.plot(years.index, 
         years.values,
         marker = "p",
         markersize=6,
         color = "red", 
         linestyle = "--",
         markerfacecolor = "black")
plt.title("Number of Movies by Year(top250)", fontsize = 16)
plt.xlabel("Years")
plt.ylabel("Counts")
plt.grid(alpha=0.5)
plt.show()
```

#### b. Seaborn kütüphanesi kullanarak, en çok film yapılan 10 türü bar grafiğiyle gösterin.

```python
df_top250.head()
```

```python
# Split movie tours from comma in the Genre column
genres = df_top250['genre'].str.split(',')
genres = genres.explode()
genres.value_counts().head(10)
```

```python
# looking at the top 10 genres of movies with a bar chart
plt.figure(facecolor = "orange",figsize = (10,7))
ax = sns.barplot(data = df_top250, x = genres.value_counts().head(10).index, y = genres.value_counts().head(10).values, palette = 'viridis')
ax.set(xlabel= 'Genres',ylabel='Count',title='Top 10 Movie Genres')
ax.set_facecolor("xkcd:mint green")
ax.bar_label(ax.containers[0]) # barlarin uzerindeki degerleri yazdirir
plt.xticks(rotation=45,color='r'); # x eksenindeki bar adlarinin acisini ve rengini ayarlar
```

#### c. Matplotlib ve Plotly kütüphanesi kullanarak, film türlerine göre filmlerin sayısını gösteren iki ayrı pasta grafik çizin.

```python
# Pie chart plot showing the number of movies by movie genres with matplotlib
plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
sns.color_palette("pastel")
plt.pie(x=genres.value_counts().values, labels=genres.value_counts().index, autopct='%1.1f%%')
plt.title('Number of Movies by Movie Genre')
plt.show()
```

```python
# Pie chart plot showing the number of movies by movie genres with plotly
import plotly.graph_objects as go

fig = go.Figure(data=[go.Pie(labels=genres.value_counts().index, values=genres.value_counts().values)])
fig.update_layout(title='Number of Movie Genres', title_x=0.5,autosize=True, width=900, height=800) 

fig.show()
```

## 9. İlişkileri görselleştirme

#### a. Seaborn kütüphanesi kullanarak, veri setindeki nümerik değişkenlerin korelasyonunu gösteren bir heatmap çizin.

```python
# Draw a heatmap showing the correlation of numerical variables in the dataset with the Seaborn library.
sns.heatmap(df_top250.corr(), annot=True, cmap="cubehelix", center = 0.6);
```

#### b. Seaborn ile oyunculardan birden fazla filmde rol alan oyuncuların rol aldıkları film sayılarına göre ilk 20 oyucuyu görselleştir.  

#### Not: Bu soru için kullanılacak data setimiz 936 satırlık data seti olacak.

```python
movies_new.sample(5)
```

```python
movies_new.shape
```

```python
# # Split actors from comma in the actors column
actor = movies_new['actors'].str.split(',')
actor = actor.explode()
actor.value_counts().head(20)
```

```python
# Visualize the top 20 actors based on the number of movies in which actors starred in more than one movie with Seaborn.
plt.style.use('seaborn-darkgrid')
plt.figure(figsize = (11,7))
ax = sns.barplot(data = df_new, y = actor.value_counts().head(20).index, x = actor.value_counts().head(20).values)
ax.set(ylabel= 'Aktors',xlabel='Count',title='Top 20 Actors')
ax.bar_label(ax.containers[0]) # barlarin uzerindeki degerleri yazdirir
plt.xticks(color='r'); # x eksenindeki bar adlarinin acisini ve rengini ayarlar
```

#### c. Plotly kütüphanesi kullanarak, Popüler 250 Filmin rating puanlarına göre hasılatını buble chart ile gösterin .

```python
df_top250.head()
```

```python
df_top250['revenue'].value_counts().index
```

```python
Show the revenue of 250 Popular Movies according to their rating points with a buble chart with the Plotly library
import plotly.express as px
bubble_data = df_top250[['rating', 'revenue', 'rank']]

fig = px.scatter(bubble_data, x='rating', y='revenue', size='revenue', 
                 hover_name='rank', title='Revenue of 250 Popular Movies by Rating Scores')
fig.update_layout(width=1100,height=600)

fig.show()
```
