```python
tips = sns.load_dataset("tips") # seaborn data setlerinin icinden tips data setini yukledik
tips.head()
```

## Scatter Plot 
* Dagilim grafikleri, farkli ozelliklerin birbiriyle olan iliskisini(kolerasyonu) goruntulemek icin kullanilir.

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip");
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips); # x ve y yi basta da verebiliriz. Positional arguman degiller
```

```python
sns.regplot(x = "total_bill", y = "tip", data = tips);
# regplot() regresyon cizgisi cizer cizginin gri golgolendirme kismi guven araligini temsil eder, hata payi
```

```python
sns.swarmplot(data=tips, x="total_bill", y="day",hue = "sex")
plt.show()
```

## figsize

```python
plt.figure(figsize = (10,5))
# sns.scatterplot(x = "total_bill", y = "tip", data = tips)
sns.scatterplot(x = tips["total_bill"], y = tips["tip"])
# bu sekilde de kullanilabilir, basta datanin adini yazdigimiz icin data+ parametresini kullanmaya gerek kalmadi
plt.show()
```

## hue

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker");# hue ile renklendirmeyi smoker sutununa gore yaptik
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "day" )
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex" )
plt.show()
```
