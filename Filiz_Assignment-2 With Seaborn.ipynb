## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill["manner_of_death"].unique()
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
kill.dropna(inplace=True)
```

```python
kill.info()
```

```python
#1.yol 
fig, ax = plt.subplots(1,1,figsize= (4.2, 3))
sns.boxplot(data = kill, x = "gender", y = "age", hue = "manner_of_death",ax = ax)  
plt.legend( title = 'manner_of_death',fontsize='x-small',title_fontsize= 'x-small')
ax.tick_params(axis = 'y',labelsize = 6)
ax.tick_params(axis = 'x',labelsize = 6)
```

```python
#2.yol
sns.boxplot(data=kill, x = "gender", y= "age", hue="manner_of_death")
plt.show()
            
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(4,3))
sns.swarmplot(data=kill, x = "gender", y= "age", hue="manner_of_death")
plt.legend( title = 'manner_of_death',fontsize='x-small',title_fontsize= 'x-small')
plt.show()
```

## 2: Plot 12 most dangerous cities

```python
most_danger_12 = kill["city"].value_counts().nlargest(12).values 
most_danger_12
```

```python
kill["city"].value_counts().nlargest(12)
```

```python
most_dangerous_cities =kill["city"].value_counts().head(13)
print(most_dangerous_cities)
```

```python
np.array(most_dangerous_cities) # Hata: dtype=int64 nasil atariz?
```

```python
most_dangerous_cities.index # Hata: Bakersfield' gelmemesi lazim!
```

```python
# With Seaborn: 
plt.figure(figsize=(8.4,5))

sns.barplot(x=most_dangerous_cities.index, y=most_dangerous_cities.values)
plt.xticks(rotation=45,color="blue",fontsize= 8)
plt.yticks(fontsize=8)
plt.title(" Most Dangerous Cities", color="red")

plt.show() 
```

```python
# with seaborn
```

```python
most_dangerous_cities =kill["city"].value_counts(ascending=False).head(13) # hata: Bakersfield  gelmemesi lazim! 
print(most_dangerous_cities)
```

```python
most_dangerous_cities.index # Hata 
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=kill, x="city",order=most_dangerous_cities.index)     # by using "most_dangerous_cities" we can create a condition
                                                                       # where we only plot those top cities
plt.title('Most Dangerous Cities', fontsize=12, color='b')
plt.tick_params(axis='x', labelsize=7, rotation=45)
plt.tick_params(axis='y', labelsize=7)
plt.xlabel("city", fontsize=7)
plt.ylabel("count", fontsize=7)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="city", data=kill, order=most_dangerous_cities.index)
plt.xticks(rotation=45,fontsize=7)
plt.yticks(fontsize=7)
plt.title("Most Dangerous Cities", color='blue')
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
most_dangerous_state =kill["state"].value_counts(ascending=False).head(20)
print(most_dangerous_state)
```

```python
# with seaborn barplot
```

## 5: Plot number of Threat Types

```python
tehdit= kill["threat_level"].value_counts()
tehdit
```

```python
# 1 .yol
plt.figure(figsize= (4,3))
grouped_data = kill.groupby("threat_level").size().reset_index(name="count")
sns.barplot(data=grouped_data, x ="threat_level", y="count")
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

plt.show()
```

```python
plt.figure(figsize=(4.2, 3))
sns.countplot(data=kill, x="threat_level")
plt.tick_params(axis='x', labelsize=7)
plt.tick_params(axis='y', labelsize=7)
plt.xlabel("threat_level", fontsize=7)
plt.ylabel("count", fontsize=7)
plt.show()
```

```python
# 2.yol
plt.figure(figsize=(4,3))

sns.barplot(x=tehdit.index, y=tehdit.values)
plt.xlabel("threat_level",fontsize=8)
plt.ylabel("count",fontsize=8)
plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8) 
plt.show()
```

```python
# with seaborn
```
