```python
# addition hue
fig, ax_tip = plt.subplots(figsize = (8,4))
ax_tip = sns.countplot(data = tips, x = "day", hue = "sex") # countplot kategorik verilerde calisir, sutunlar getirr barplottan farki aralarda bosluk olmasi
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x() +0.15, p.get_height() + 1));
```

```python
fig, ax= plt.subplots(figsize = (12,6))
ax = sns.countplot(x = "day",data = tips, hue = "sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.15, p.get_height() + 1));
# get_height() ile sutunun ustune yaz demek, bos kullanbiliyoruz(defaultlari kullaniyor) veya yanina deger girebiliyoruz
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
ax = sns.countplot(x = "day",  data = tips, hue = "sex");
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+0.7))
    ax.text(p.get_x()+0.07, p.get_height()-4, "%"+str(round(p.get_height()/tips.day.count()*100, 2))) # yuzde degerlerini hesaplatip str a cevirdik  
plt.legend(loc = "upper left");
#ax.text ile bu sefer barin icine yuzdelik degerini yazdirdim
#yuzdelik hesaplamasini yaptirdim, sonuna da 2 yazdim ki virgulden sonra 2 basamak getirsin
#roundunu aldirdim, integer cikan degeri de str'ye cevirdim, basina da % isaretini str olarak koydum.
#height'a -4 verince de barin icine yazdi 
```

## Bar Plot
ci = confidance interyal(Guven araligi %95) %5 hata yapabiliriz. :))

Barlar, verilen hizalama ile x'te konumlandırılır. Boyutları yükseklik ve genişlik ile verilir. Dikey taban çizgisi alttır (varsayılan 0).

Birçok parametre ya tüm barlara uygulanan tek bir değer ya da her bar için bir tane olmak üzere bir dizi değer alabilir.



```python
sns.barplot(data = tips, x = "sex", y = "tip");
```

```python
# add mean estimator
sns.barplot(x="sex", y="tip", data = tips, estimator = np.mean,ci = "sd");
# cizgi ne kadar uzun ise verinin standart sapmasi o kadar fazladir.
```

```python
# add sum() estimator
sns.barplot(x="sex", y="tip", data = tips, estimator = np.sum,ci = "sd");
# cizgi ne kadar uzun ise verinin standart sapmasi o kadar fazladir.
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x = "day", data = tips)
for p in ax.patches:
    ax.annotate("("+str(p.get_height())+")" +
                "%" + str(round((p.get_height())/ tips.day.count(), 2)*100), (p.get_x() + 0.05, p.get_height()-10));
    
```
