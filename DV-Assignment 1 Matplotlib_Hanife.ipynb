## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd
import matplotlib as mpl #versiyon kontrol, default degerlere donme vs icin
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#%matplotlib notebook
import warnings; # uyari mesajlarini almamak icin
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(11, dtype="float")
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (4.2,3))
plt.plot(m,E,linewidth= 4,color="b" ) # linewidth= 4 cizgilerin kalinligini ayarlar, lisaca lw=4 seklinde yazabiliriz
plt.xlabel("M in g", fontsize = 7)
plt.ylabel("E in j", fontsize = 7)
plt.title("E = mc^2",fontsize = 8);
```

```python
plt.figure(figsize = (5.7,4))
plt.plot(m,E,lw= 2,color="orange" ) # linewidth= 4 cizgilerin kalinligini ayarlar, lisaca lw=4 seklinde yazabiliriz
plt.xlabel("M in g", fontsize = 7)
plt.ylabel("E in j", fontsize = 7)
plt.title("E = mc^2",fontsize = 8);
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.figure(figsize = (4.2,3))
plt.semilogy(m,E,lw= 2,color="green" ) #Bu , günlük ölçeklemeye ek olarak y eksenini değiştiren ince bir sarmalayıcıdır .
plt.xlim([0,10])

plt.xlabel("M in g", fontsize = 7)
plt.ylabel("E in j", fontsize = 7)

plt.grid(which='both', axis="y")
plt.title("E = mc^2",fontsize = 8);
```

## Aşağıdaki değişkenlere göre ilgili plotlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize = (5.7,4.5))
plt.plot(july16_2007,lw = 1,label = "july16_2007")
plt.plot(july16_2020,lw= 1,color="orange",label = "july16_2020" )
plt.legend()
```

```python
plt.figure(figsize = (5.7,4.5))
plt.plot(july16_2007,lw = 1,label = "july16_2007")
plt.plot(july16_2020,lw= 1,color="orange",label = "july16_2020" )
plt.legend(bbox_to_anchor=(0.63,0,0.6,0.5),fontsize= 'x-small');
# bbox_to_anchor=(0.0,0.0,0.9,0.9) #ilk iki deger ile figure de gezebilirken son iki degerde axes te gezeriz..ince ayar yapar
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
mpl.rcParams.update(mpl.rcParamsDefault) ## Grafiklerdeki tüm modifikasyonları default değerlere döndürdük
fig, ax = plt.subplots(2,1,figsize= (7.2, 5))

ax[0].plot(labels, july16_2007,lw=2)
ax[0].set_title("july16_2007") # setting title of subplot
ax[0].tick_params(axis='x', labelsize=7.5)
ax[0].tick_params(axis='y', labelsize=7.5)

ax[1].plot(labels, july16_2020,lw=2)
ax[1].set_title("july16_2020") 
ax[1].tick_params(axis='x', labelsize=7.5) # tick_params ile eksenlere istedigimiz ozellikleri uygulatabiliriz
ax[1].tick_params(axis='y', labelsize=7.5)

fig.tight_layout()
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots(figsize=(8.7, 5.7))

ax2 = ax1.twinx() # X eksenini paylaşan bir ikiz Eksen oluşturduk

ax1.plot(labels, july16_2007, color='blue', label='July 16, 2007', lw=2)
ax1.set_ylabel("July 16 2007", color = "b", fontsize = 14)

ax2.plot(labels, july16_2020, color='red', label='July 16, 2020', lw=2)
ax2.set_ylabel('July 16 2020', color = "r", fontsize = 14)

ax1.tick_params(axis='x', labelsize=7) # tick_params ile eksenlere istedigimiz ozellikleri uygulatabiliriz
ax1.tick_params(axis='y', labelsize=7)
ax2.tick_params(axis='y', labelsize=7)

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8.7, 5.7))

ax2 = ax1.twinx() # X eksenini paylaşan bir ikiz Eksen oluşturduk

ax1.plot(labels, july16_2007, color='blue', label='July 16, 2007', lw=2)
ax1.set_ylabel("July 16 2007", color = "b", fontsize = 14)


ax2.plot(labels, july16_2020, color='red', label='July 16, 2020', lw=2)
ax2.set_ylabel('July 16 2020', color = "r", fontsize = 14)

ax2.spines['right'].set_color('red') # sag eksenin rengini ayarladik
ax2.spines['left'].set_color('blue') # sol eksenin rengini ayarladik
ax2.spines['left'].set_linewidth(1.5) # sol eksenin kalinligini ayarladik
ax2.spines['right'].set_linewidth(1.5) # sag eksenin kalinlgini ayarladik

ax1.tick_params(axis='x', labelsize=7) # tick_params ile eksenlere istedigimiz ozellikleri uygulatabiliriz
ax1.tick_params(axis='y', labelsize=7)
ax2.tick_params(axis='y', labelsize=7)

plt.show()
```
